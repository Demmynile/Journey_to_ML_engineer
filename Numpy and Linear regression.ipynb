```python
def create_vector_vector_multiplication(u , v):
  assert u.shape[0] == v.shape[0]

  n = u.shape[0]

  result = 0.0

  for i in range(n):
    result = result + u[i] + v[i]

  return result
```

```python
create_vector_vector_multiplication(u,v)
```

```python
def create_matrix_vector_multiplication(U , v):
    assert U.shape[0] == v.shape[1]

    num_rows = U.shape[0]

    result = np.zeros(num_rows)

    for i in range(num_rows):
      result[i] = result + create_vector_vector_multiplication(U[i] , v)

    return result
```
